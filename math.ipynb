```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
# plt.rc('text', usetex=True)
# plt.rc('font', family='serif')

t = np.arange(-5., 5., 0.05)
s1 = 50 * np.cos(2 * np.pi * t)
s2 = np.exp(-t) * np.cos(2 * np.pi * t)
s3 = 100 * np.exp(-t*t)

# plt.subplot(311)
plt.plot(t, s1, 'r', lw=1, label='$50 cos(2 \pi t)$') # '50 * cos(2 * pi * t)'
plt.grid(True)
plt.ylim(-150, 150)
plt.legend(loc='lower left')

# plt.subplot(312)
plt.plot(t, s2, 'g', lw=2, label='$e^{-t} cos(2 \pi t)$') # 'e^(-t) * cos(2 * pi * t)'
plt.grid(True)
plt.ylim(-150, 150)
plt.legend(loc='lower left')

# plt.subplot(313)
plt.plot(t, s3, 'b--', lw=2, label='$100 e^{-t^2}$') # '100 * e^(-t * t)'
plt.grid(True)
plt.ylim(-150, 150)
plt.legend(loc='lower left')

plt.show()
```
